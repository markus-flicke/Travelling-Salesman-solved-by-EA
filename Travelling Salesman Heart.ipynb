## 2.0 Travelling Salesman

```python
from Path import Path
from TravellingSalesmanProblem import TravellingSalesmanProblem
from random import random
from time import time
import matplotlib.pyplot as plt
```

```python
points = [(0.5,0), (0.53, 0.45), (0.47, 0.45), (0.65,0.4), (0.351,0.45), (0.649,0.45), (0.35,0.4), (0.64,0.5), (0.36,0.5),(0.6,0.55), (0.4,0.55),(0.5,0.4), (0.57,0.52), (0.43,0.52)]
plt.scatter([x[0] for x in points], [y[1] for y in points], marker = 'x')
```

```python
t = TravellingSalesmanProblem(points=points, pop_size=20)
t.next_generation()
t.get_fittest().plot()
```

```python
start = time()
fitnesses = []
while time() < start + 3:
    t.next_generation()
    fitnesses.append(t.fitness(t.get_fittest()))
plt.plot(fitnesses)
```

```python
t.get_fittest().plot(color = 'red')
```

## Mutation  
Should switch exactly two edges

```python
N=5
points = [(random(), random()) for i in range(N)]
p = Path(points=points)
p.plot()
```

```python
p = p.mutate()
p.plot(color='orange')
```

## Recombination
Child should contain only edges that exist in parents

```python
N = 6
points = [(random(), random()) for i in range(N)]
mum = Path(points=points)
dad = Path(points=points)
child = mum.recombine(dad)
mum.plot(color = 'red')
```

```python
dad.plot()
```

```python
child.plot(color='green')
```
