## 2.0 Travelling Salesman

```python
from Path import Path
from TravellingSalesmanProblem import TravellingSalesmanProblem
from random import random
from time import time
import matplotlib.pyplot as plt
```

```python
t = TravellingSalesmanProblem(N=100, pop_size=20)
t.next_generation()
t.get_fittest().plot()
```

```python
start = time()
fitnesses = []
while time() < start + 30:
    t.next_generation()
    fitnesses.append(t.fitness(t.get_fittest()))
plt.plot(fitnesses)
```

```python
gradients = []
for i in range(len(fitnesses)-1):
    gradient = fitnesses[i+1] - fitnesses[i]
    gradients.append(gradient)
plt.subplots(1,1,figsize=(20,8))
plt.scatter(range(len(gradients)),gradients, marker = 'x')
```

```python
import pandas as pd
pd.Series(gradients)
```

## Mutation  
Should switch exactly two edges

```python
N=5
points = [(random(), random()) for i in range(N)]
p = Path(points=points)
p.plot()
```

```python
p = p.mutate()
p.plot(color='orange')
```

## Recombination
Child should contain only edges that exist in parents

```python
N = 6
points = [(random(), random()) for i in range(N)]
mum = Path(points=points)
dad = Path(points=points)
child = mum.recombine(dad)
mum.plot(color = 'red')
```

```python
dad.plot()
```

```python
child.plot(color='green')
```
